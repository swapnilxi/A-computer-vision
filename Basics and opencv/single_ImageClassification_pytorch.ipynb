## Single label Image Classification

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
from PIL import Image
```

```python
InputImg= Image.open("/content/drive/MyDrive/Colab Notebooks/ComputerVision/ImageData/Pizza.jpg")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(InputImg)
```

```python
from torchvision import transforms
transform = transforms.Compose([
 transforms.Resize(256),
 transforms.CenterCrop(224),
 transforms.ToTensor(),
 transforms.Normalize(
 mean=[0.485, 0.456, 0.406],
 std=[0.229, 0.224, 0.225]
 )])
```

## Transforming the image

```python
InputImg_t = transform(InputImg)
```

```python
print(InputImg_t.shape)
```

```python
#plt.imshow(InputImg_t). --> won't work with pytorch tensor
plt.imshow(InputImg_t.permute(1, 2, 0))
# permute rearranges the dimensions of the tensor.
#Here, 1, 2, 0 changes the order from (C, H, W) to (H, W, C)
# which is the format matplotlib expects.
```

```python
import torch

InputImg_bt = torch.unsqueeze(InputImg_t, 0);
print(InputImg_bt.shape)
```

```python
print(InputImg_bt.shape)
```

#using model for image classification


```python
from torchvision import models
dir(models)
```

```python
resnet = models.resnet18(pretrained=True)
resnet.eval()
```

```python
output = resnet(InputImg_bt)
```

```python
alexnet = models.alexnet(pretrained=True)
alexnet.eval()
```

```python
output = alexnet(InputImg_bt)
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/ComputerVision/udemy/Modern_computer_Vision/ImageClassification/imagenet1000Classes.txt') as classesfile:
  ImageNetclasses = [line.strip() for line in classesfile.readlines()]

_, predicted = torch.max(output, 1)
percentage = torch.softmax(output, dim=1)[0] * 100
print(ImageNetclasses[predicted[0]], percentage[predicted[0]].item())
```
